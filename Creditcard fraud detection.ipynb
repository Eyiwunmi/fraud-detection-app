```python
from sklearn.preprocessing import StandardScaler


def scale_creditcard_data(df):
    """
    Standardize the feature columns in the credit card fraud dataset.

    This function separates the dataset into features (X) and target (y),
    then applies StandardScaler to normalize all numeric feature columns
    so that each has a mean of 0 and a standard deviation of 1.
    This ensures that features like 'Amount' and PCA components (V1‚ÄìV28)
    are on the same scale, improving model performance and convergence.

    Parameters:
        df (pd.DataFrame): The full credit card fraud dataset, including the 'Class' column.

    Returns:
        X_scaled (pd.DataFrame): Standardized feature DataFrame.
        y (pd.Series): Target variable representing fraud (1) and non-fraud (0).
    """
    # Separate features (X) and target (y)
    X = df.drop('Class', axis=1)
    y = df['Class']

    # Create the scaler object
    scaler = StandardScaler()

    # Fit the scaler on features and transform
    X_scaled = scaler.fit_transform(X)

    # Convert the scaled data back to a DataFrame
    X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

    print("‚úÖ Features successfully standardized using StandardScaler!")
    return X_scaled, y

# Example usage
X_scaled, y = scale_creditcard_data(df)
```

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score, train_test_split


def train_with_cv_and_smote(df):
    """
    Train and evaluate a Random Forest model using cross-validation 
    with SMOTE applied safely inside each fold (to prevent data leakage).
    
    Parameters
    ----------
    df : pd.DataFrame
        The credit card fraud dataset with a 'Class' column.
    
    Returns
    -------
    None
    """
    # Split into features and labels
    X = df.drop('Class', axis=1)
    y = df['Class']

    # Train/test split (final evaluation holdout)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )

    # Define pipeline (SMOTE inside)
    pipeline = Pipeline([
        ('smote', SMOTE(random_state=42)),
        ('model', RandomForestClassifier(
            n_estimators=100,
            class_weight='balanced',
            random_state=42
        ))
    ])

    # Stratified cross-validation
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # Evaluate using F1 score (since fraud detection = imbalanced)
    f1_scores = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring='f1', n_jobs=-1)

    print(f"Cross-Validation F1 Scores: {f1_scores}")
    print(f"Mean F1 Score: {f1_scores.mean():.4f}")

    # Fit on entire training data (using SMOTE inside pipeline)
    pipeline.fit(X_train, y_train)

    # Evaluate on unseen test data (no SMOTE applied)
    test_f1 = pipeline.score(X_test, y_test)
    print(f"\nF1 Score on Test Set: {test_f1:.4f}")

    return pipeline, X_train, X_test, y_train, y_test
```

```python
from sklearn.metrics import classification_report

pipeline, X_train, X_test, y_train, y_test = train_with_cv_and_smote(df)
y_pred = pipeline.predict(X_test)
print("\nClassification Report on Test Data:")
print(classification_report(y_test, y_pred))
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    roc_auc_score,
    precision_score,
    recall_score,
    f1_score
)
import pandas as pd

def compare_models(X_train, X_test, y_train, y_test):
    """
    Train and compare multiple models for credit card fraud detection.
    
    Models compared:
        - Logistic Regression
        - Random Forest
        - XGBoost
        
    For each model, the following metrics are displayed:
        - Precision (Class = 1)
        - Recall (Class = 1)
        - F1-score (Class = 1)
        - ROC-AUC score
        - Confusion Matrix
    
    Parameters:
        X_train, X_test, y_train, y_test : pd.DataFrame or np.array
            Training and testing data.
    
    Returns:
        pd.DataFrame: Summary table of model performance.
    """
    
    models = {
        "Logistic Regression": LogisticRegression(max_iter=1000, class_weight='balanced'),
        "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced'),
        "XGBoost": XGBClassifier(scale_pos_weight=99, random_state=42, use_label_encoder=False, eval_metric='logloss')
    }

    results = []

    for name, model in models.items():
        print(f"\n=== Training {name} ===")
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_pred_proba = model.predict_proba(X_test)[:, 1]

        # Metrics
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, y_pred_proba)
        cm = confusion_matrix(y_test, y_pred)

        print("\nConfusion Matrix:\n", cm)
        print("\nClassification Report:\n", classification_report(y_test, y_pred))
        print(f"ROC-AUC Score: {roc_auc:.4f}")

        # Save results
        results.append({
            "Model": name,
            "Precision": precision,
            "Recall": recall,
            "F1-Score": f1,
            "ROC-AUC": roc_auc
        })

    # Create comparison table
    results_df = pd.DataFrame(results).sort_values(by="F1-Score", ascending=False)
    print("\n=== üìä Model Performance Comparison ===")
    print(results_df)

    return results_df

# Example usage:
results_df = compare_models(X_train, X_test, y_train, y_test)
```

```python
# =======================
# ‚ö° FAST XGBoost Fine-Tuning & Evaluation
# =======================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import (
    confusion_matrix, precision_score, recall_score,
    f1_score, roc_auc_score
)

def fast_tune_xgb(X_train, X_test, y_train, y_test):
    """
    Quickly tune and evaluate an XGBoost model on the fraud dataset.
    Uses a small hyperparameter grid for speed and includes threshold tuning
    and feature importance visualization.

    Parameters
    ----------
    X_train, X_test, y_train, y_test : DataFrame/Series
        Split data.

    Returns
    -------
    dict : Model performance results.
    """

    # ‚öôÔ∏è Compact but effective parameter grid for fast search
    grid_params = {
        'n_estimators': [100, 200],
        'max_depth': [3, 5],
        'learning_rate': [0.05, 0.1],
        'scale_pos_weight': [99],  # Handle class imbalance
        'subsample': [0.8],
        'colsample_bytree': [0.8]
    }

    xgb = XGBClassifier(
        random_state=42,
        tree_method='hist',   # ‚úÖ fastest algorithm for large data
        eval_metric='logloss',
        use_label_encoder=False
    )

    cv = StratifiedKFold(n_splits=2, shuffle=True, random_state=42)

    print("üöÄ Running a quick grid search for XGBoost...")
    grid = GridSearchCV(
        xgb, grid_params, scoring='f1', cv=cv,
        n_jobs=-1, verbose=2
    )
    grid.fit(X_train, y_train)

    best_model = grid.best_estimator_
    print("\n‚úÖ Best Parameters:", grid.best_params_)

    # Predictions & probabilities
    y_proba = best_model.predict_proba(X_test)[:, 1]
    y_pred = (y_proba >= 0.5).astype(int)

    # Default threshold metrics
    metrics_default = {
        'precision': precision_score(y_test, y_pred, zero_division=0),
        'recall': recall_score(y_test, y_pred, zero_division=0),
        'f1': f1_score(y_test, y_pred, zero_division=0),
        'roc_auc': roc_auc_score(y_test, y_proba)
    }

    print("\nüìä Default threshold (0.5) metrics:")
    print(metrics_default)

    # Threshold tuning (maximize F1)
    thresholds = np.linspace(0.1, 0.9, 9)
    f1_scores = []
    for t in thresholds:
        y_t = (y_proba >= t).astype(int)
        f1_scores.append(f1_score(y_test, y_t, zero_division=0))

    best_idx = np.argmax(f1_scores)
    best_threshold = thresholds[best_idx]

    print(f"\nüéØ Best threshold: {best_threshold:.2f} (F1 = {f1_scores[best_idx]:.4f})")

    # Confusion matrix
    cm = confusion_matrix(y_test, (y_proba >= best_threshold).astype(int))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix (threshold={best_threshold:.2f})')
    plt.xlabel('Predicted'); plt.ylabel('Actual')
    plt.show()

    # Feature importance
    fi = pd.DataFrame({
        'Feature': X_train.columns,
        'Importance': best_model.feature_importances_
    }).sort_values('Importance', ascending=False)

    plt.figure(figsize=(8, 6))
    sns.barplot(x='Importance', y='Feature', data=fi.head(10))
    plt.title('Top 10 Feature Importances (XGBoost)')
    plt.show()

    return {
        'best_model': best_model,
        'best_params': grid.best_params_,
        'metrics_default': metrics_default,
        'best_threshold': best_threshold
    }

# Example usage
results = fast_tune_xgb(X_train, X_test, y_train, y_test)
```

```python
from sklearn.metrics import (
    classification_report, confusion_matrix,
    precision_score, recall_score, f1_score, roc_auc_score
)
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


def evaluate_unseen_data(results, X_unseen, y_unseen):
    """
    Evaluate the tuned XGBoost model on unseen (new) data.

    Parameters
    ----------
    results : dict
        Output dictionary returned from `fast_tune_xgb()`. 
        Must contain 'best_model' and 'best_threshold'.
    X_unseen : pd.DataFrame
        Feature set of unseen (new) data.
    y_unseen : pd.Series
        True labels for the unseen data.

    Returns
    -------
    dict
        Evaluation metrics on unseen data, including:
        - precision
        - recall
        - f1-score
        - roc_auc
    """

    # Extract model and threshold
    best_model = results['best_model']
    best_threshold = results['best_threshold']

    # Predict probabilities
    y_proba = best_model.predict_proba(X_unseen)[:, 1]

    # Apply tuned threshold
    y_pred_final = (y_proba >= best_threshold).astype(int)

    # ---- Display Results ----
    print(f"\nüìã Classification Report (threshold={best_threshold:.2f}):")
    print(classification_report(y_unseen, y_pred_final, digits=4))

    # Confusion Matrix
    cm = confusion_matrix(y_unseen, y_pred_final)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Oranges', cbar=False)
    plt.title(f'Confusion Matrix on Unseen Data (threshold={best_threshold:.2f})')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Compute metrics
    metrics = {
        'precision': precision_score(y_unseen, y_pred_final, zero_division=0),
        'recall': recall_score(y_unseen, y_pred_final, zero_division=0),
        'f1': f1_score(y_unseen, y_pred_final, zero_division=0),
        'roc_auc': roc_auc_score(y_unseen, y_proba)
    }

    print("\nüìä Unseen Data Metrics:")
    for k, v in metrics.items():
        print(f"{k}: {v:.4f}")

    return metrics


# Example usage:
unseen_metrics = evaluate_unseen_data(results, X_unseen, y_unseen)
```

```python

def flag_transactions(model, X_data, high_thresh=0.9, low_thresh=0.7):
    """
    Flags transactions based on fraud probability scores.

    Parameters
    ----------
    model : trained model
        Your trained XGBoost or other classifier (must have predict_proba()).
    X_data : pd.DataFrame
        Transaction features to score.
    high_thresh : float
        Probability threshold above which transactions are automatically flagged as fraud.
    low_thresh : float
        Probability range for human review (between low_thresh and high_thresh).

    Returns
    -------
    pd.DataFrame
        DataFrame with fraud probabilities and flagging decisions:
        - 'fraud_probability'
        - 'fraud_flag' : 'FRAUD', 'REVIEW', or 'CLEAR'
    """

    # Get fraud probability scores (probability that each transaction is fraud)
    probs = model.predict_proba(X_data)[:, 1]

    # Create a results DataFrame
    results = X_data.copy()
    results["fraud_probability"] = probs

    # Apply the 3-tier decision rule
    conditions = [
        results["fraud_probability"] >= high_thresh,          # Very sure it's fraud
        (results["fraud_probability"] >= low_thresh) & 
        (results["fraud_probability"] < high_thresh),         # Uncertain zone
        results["fraud_probability"] < low_thresh              # Likely safe
    ]

    choices = ["FRAUD", "REVIEW", "CLEAR"]

    results["fraud_flag"] = np.select(conditions, choices, default="CLEAR")

    # Print summary of outcomes
    print("\nüîç Transaction flag summary:")
    print(results["fraud_flag"].value_counts())

    return results
```
